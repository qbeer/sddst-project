```python
def display_and_save_figure(path, title, xlim=10):
    df = pd.read_csv(path, sep=' ', header=None)
    df.columns = ['simulation time', 'number of successful steps',
        'number of failed steps', 'worst error ratio squared',
        'average speed of the dislocations', 'cutoff (used in the semi implicit scheme)',
        'order parameter', 'value of the external stress',
        'computation time between the last two successful steps', 'accumulated strain',
        'average v2', 'energy of the system']
    
    time = np.log(df['simulation time'])
    energy = df['energy of the system']
    speed = np.log(df['average speed of the dislocations'])
    speed2 = np.log(df['average v2'])
    
    
    plt.figure(figsize=(12, 9))
    plt.subplot("221")
    plt.title("Dissipated energy - Time")
    plt.plot(time, energy, 'b.--')
    plt.xlim(-.1, xlim)
    plt.ylabel("E")
    plt.xlabel("ln(t)")
    plt.subplot("222")
    plt.title("Average speed - Time")
    plt.plot(time, speed, 'r.--')
    plt.xlim(-.1, xlim)
    plt.ylabel("ln(<v>)")
    plt.xlabel("ln(t)")
    plt.subplot("223")
    plt.title("Average speed squared - Time")
    plt.plot(time, speed2, 'g.--')
    plt.xlim(-.1, xlim)
    plt.ylabel("ln(<v*v>)")
    plt.xlabel("ln(t)")

    plt.tight_layout()
    plt.savefig("%s.png" % title)
```
