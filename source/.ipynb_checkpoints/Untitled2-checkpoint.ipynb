```python
def plot_locs(path, title, num):
    locs = pd.read_csv(path, header=None, sep=' ')
    locs.columns = ['x', 'y', 'b']
    
    x = locs['x'].values
    y = locs['y'].values
    
    plt.figure(figsize=(7,7))

    plt.plot(x, y, 'bo', label='N = %d analytic' % num)
    plt.plot(x_, y_, 'rx', label='Original')
    plt.title(title)
    plt.xlim(-0.5, 0.5)
    plt.ylim(-0.5, 0.5)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='upper right')
    
    plt.tight_layout()
    plt.savefig("%s.png" % title)
```

```python
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib import ticker
from sklearn.preprocessing import normalize

img = pd.read_csv('../build/field.dat', header=None, sep=";")
img = img.dropna(axis=1)

#img = normalize(np.array(img))

print(np.mean(np.min(img)), np.mean(np.max(img)))

plt.figure(figsize=(7, 7))
ax = plt.subplot(111)

plt.title("ELTE stress field", fontdict={"fontsize": 14})
plt.xlabel("x", fontdict={"fontsize": 13})
plt.ylabel("y", fontdict={"fontsize": 13})
    
plt.xticks(np.arange(0, img.shape[0], img.shape[0] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
plt.yticks(np.arange(0, img.shape[1], img.shape[1] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
 
pcm = ax.pcolor(img,
                 norm=colors.SymLogNorm(linthresh=0.4, linscale=0.4,
                                              vmin=-1.5, vmax=1.5),
                 cmap=plt.cm.Spectral_r)
    
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="7%", pad="2%")

cb = plt.colorbar(pcm, cax=cax)
tick_locator = ticker.MaxNLocator(nbins=5)
cb.locator = tick_locator
cb.update_ticks()

plt.savefig('elte_stress_field.png', dpi=120)
plt.plot()
```

```python
x, y = np.linspace(-0.5, 0.5, 200), np.linspace(-0.5, 0.5, 200)
xx, yy = np.meshgrid(x, y)
```

```python
G = 27.7e09 #Pa
v = 0.334

tau = G*1/(2*np.pi*(1-v))*xx*(xx**2 - yy**2)/(xx**2 + yy**2)**2
```

```python
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib import ticker
from sklearn.preprocessing import StandardScaler

img = tau

img = StandardScaler().fit_transform(np.array(img)).T

print(np.mean(np.min(img)), np.mean(np.max(img)))

plt.figure(figsize=(7, 7))
ax = plt.subplot(111)

plt.title("TAU_XY flipped", fontdict={"fontsize": 14})
plt.xlabel("x", fontdict={"fontsize": 13})
plt.ylabel("y", fontdict={"fontsize": 13})
    
plt.xticks(np.arange(0, img.shape[0] + 1, (img.shape[0] + 1) // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
plt.yticks(np.arange(0, img.shape[1] + 1, (img.shape[1] + 1) // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
 
pcm = ax.pcolor(img,
                 norm=colors.SymLogNorm(linthresh=0.7, linscale=0.7,
                                              vmin=-2, vmax=2),
                 cmap=plt.cm.Spectral_r)
    
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="7%", pad="2%")

cb = plt.colorbar(pcm, cax=cax)
tick_locator = ticker.MaxNLocator(nbins=7)
cb.locator = tick_locator
cb.update_ticks()

plt.savefig('elte_stress_field_TAU_XY_flipped.png', dpi=120)
plt.plot()
```

```python
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib import ticker
from sklearn.preprocessing import StandardScaler, normalize

img = pd.read_csv('../build/field.dat', header=None, sep=";")
img = img.dropna(axis=1)
img = img.values[:200, :200]

tau_ = normalize(tau_)

plt.figure(figsize=(7, 7))
ax = plt.subplot(111)

plt.title("ELTE stress field - TAU_XY", fontdict={"fontsize": 14})
plt.xlabel("x", fontdict={"fontsize": 13})
plt.ylabel("y", fontdict={"fontsize": 13})
    
plt.xticks(np.arange(0, img.shape[0], img.shape[0] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
plt.yticks(np.arange(0, img.shape[1], img.shape[1] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
 
pcm = ax.pcolor(tau_,
                 norm=colors.SymLogNorm(linthresh=0.001,
                                        linscale=0.001,
                                              vmin=-1, vmax=1),
                 cmap=plt.cm.Spectral_r)
    
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="7%", pad="2%")

cb = plt.colorbar(pcm, cax=cax)
tick_locator = ticker.MaxNLocator(nbins=9)
cb.locator = tick_locator
cb.update_ticks()

plt.plot()
```

```python
dfs = [pd.read_csv('../build/outputs/dat%d.log' % i, sep=' ', header=None) for i in range(1, 11)]

print(len(dfs))

for df in dfs:
    df.columns = ['simulation time', 'number of successful steps',
        'number of failed steps', 'worst error ratio squared',
        'average speed of the dislocations', 'cutoff (used in the semi implicit scheme)',
        'order parameter', 'value of the external stress',
        'computation time between the last two successful steps', 'accumulated strain',
        'average v2', 'energy of the system']

    
times = [df['simulation time'] for df in dfs]
energys = [df['energy of the system'] for df in dfs]
speeds = [df['average speed of the dislocations'] for df in dfs]
speed2s = [df['average v2'] for df in dfs]

colors = ['r', 'g', 'b', 'c', 'y', 'm', 'k', '#eeefff',  '#040fff',  '#ee223f',  '#001212', '#123133']

plt.figure(figsize=(16, 11))
for i in range(0, 10):
    plt.subplot("221")
    plt.title("Dissipated energy - Time")
    plt.scatter(times[i], energys[i], c=colors[i], label='ELTE %d' % (i+1))
    plt.xlim(-1, 100)
    plt.legend(loc='upper right')
    plt.xticks(np.arange(0, 100, 20))
    plt.xlabel("ln(t)")
    plt.subplot("222")
    plt.title("Average speed - Time")
    plt.loglog(times[i], speeds[i], c=colors[i], label='ELTE %d' % (i+1))
    plt.xlim(0, 1000)
    plt.ylim(0.00000001, 10000)
    plt.legend(loc='upper right')
    plt.xlabel("ln(t)")
    plt.subplot("223")
    plt.title("Average speed squared - Time")
    plt.loglog(times[i], speed2s[i], c=colors[i],  label='ELTE %d' % (i+1))
    plt.xlim(0, 1000)
    plt.ylim(0.000000001, 100000)
    plt.xlabel("ln(t)")

plt.tight_layout()
plt.legend(loc='upper right')

plt.savefig('all_elte.png')
plt.show()
```

```python
dfs = [pd.read_csv('../build/outputs/dat%d.log' % i, sep=' ', header=None) for i in range(1, 11)]

for df in dfs:
    df.columns = ['simulation time', 'number of successful steps',
        'number of failed steps', 'worst error ratio squared',
        'average speed of the dislocations', 'cutoff (used in the semi implicit scheme)',
        'order parameter', 'value of the external stress',
        'computation time between the last two successful steps', 'accumulated strain',
        'average v2', 'energy of the system']

    
times = np.array([df['simulation time'].values[:641] for df in dfs])
energys = np.array([df['energy of the system'].values[:641] for df in dfs])
speeds = np.array([df['average speed of the dislocations'].values[:641] for df in dfs])
speed2s = np.array([df['average v2'].values[:641] for df in dfs])

times = [np.mean(times, axis=0)]
energys = [np.mean(energys, axis=0)]
speeds = [np.mean(speeds, axis=0)]
speed2s = [np.mean(speed2s, axis=0)]

colors = [ 'k', '#eeefff',  '#040fff',  '#ee223f',  '#001212', '#123133']

plt.figure(figsize=(16, 11))
for i in range(0, 1):
    plt.subplot("221")
    plt.title("Dissipated energy - Time")
    plt.scatter(times[i], energys[i], c=colors[i], label='ELTE averged')
    plt.xlim(-1, 60)
    plt.legend(loc='upper right')
    plt.xticks(np.arange(0, 100, 20))
    plt.xlabel("ln(t)")
    plt.subplot("222")
    plt.title("Average speed - Time")
    plt.loglog(times[i], speeds[i], c=colors[i], label='ELTE averaged')
    plt.xlim(0, 1000)
    plt.ylim(0.01, 1000)
    plt.legend(loc='upper right')
    plt.xlabel("ln(t)")
    plt.subplot("223")
    plt.title("Average speed squared - Time")
    plt.loglog(times[i], speed2s[i], c=colors[i],  label='ELTE averaged')
    plt.xlim(0, 1000)
    plt.ylim(0.1, 1000000000)
    plt.xlabel("ln(t)")

plt.tight_layout()
plt.legend(loc='upper right')

plt.savefig('all_elte_averaged.png')
plt.show()
```
