```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = pd.read_csv("elte_stress_field/field.dat", sep=";", header=None)
```

```python
img = img.dropna(axis=1)
img.head(n=10)
```

```python
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib import ticker
```

```python
#img = normalize(img)
```

```python
plt.figure(figsize=(7, 7))
ax = plt.subplot(111)

plt.title("ELTE stress field", fontdict={"fontsize": 14})
plt.xlabel("x", fontdict={"fontsize": 13})
plt.ylabel("y", fontdict={"fontsize": 13})
    
plt.xticks(np.arange(0, img.shape[0], img.shape[0] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
plt.yticks(np.arange(0, img.shape[1], img.shape[1] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
 

pcm = ax.pcolor(img,
                 norm=colors.SymLogNorm(linthresh=0.2, linscale=0.2,
                                              vmin=np.min(img)[0], vmax=np.max(img)[0]),
                 cmap=plt.cm.Spectral_r)
    
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="7%", pad="2%")

cb = plt.colorbar(pcm, cax=cax)
tick_locator = ticker.MaxNLocator(nbins=5)
cb.locator = tick_locator
cb.update_ticks()

plt.savefig('elte_stress_field.png', dpi=120)

plt.show()
```

```python
import os

files = sorted(os.listdir('./analytic_stress_field/'))
files
```

```python
images = [pd.read_csv("analytic_stress_field/" + file,
                      sep=";",
                      header=None) for file in files]

normalized_images = []

for i in range(len(images)):
    
    normalized_images.append(images[i].dropna(axis=1))
    
    plt.figure(figsize=(7, 7))
    ax = plt.subplot(111)
    
    plt.title("N = %s - analytic fields" % files[i][(5 if files[i][5] != '0' else 6):7], fontdict={"fontsize": 14})
    plt.xlabel("x", fontdict={"fontsize": 13})
    plt.ylabel("y", fontdict={"fontsize": 13})
    
    plt.xticks(np.arange(0, normalized_images[i].shape[0], normalized_images[i].shape[0] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
    plt.yticks(np.arange(0, normalized_images[i].shape[1], normalized_images[i].shape[1] // 4),
               ('-0.5', '-0.25', '0', '0.25', '0.5'), size=12)
    
    pcm = ax.pcolor(normalized_images[i],
                    norm=colors.SymLogNorm(linthresh=0.095, linscale=0.095,
                    vmin=np.min(normalized_images[i])[0],
                    vmax=np.max(normalized_images[i][0])),
                    cmap=plt.cm.Spectral_r)
    
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="7%", pad="2%")

    cb = plt.colorbar(pcm, cax=cax)
    tick_locator = ticker.MaxNLocator(nbins=3)
    cb.locator = tick_locator
    cb.update_ticks()
    
    plt.savefig('stress_field_%s.png' % files[i], dpi=120)
    plt.show()
```

```python
fig, axes = plt.subplots(3, 3, sharex=True, sharey=True, squeeze=False, figsize=(18, 18))

for i in range(3):
    for j in range(3):

        axes[i, j].set_visible(False)

for i in range(0, 3):
    for j in range(i+1, 4):
        
        axes[i, j - 1].set_title("(%d) field - (%d) field" % (i + 1, j + 1))
        
        axes[i, j - 1].set_visible(True)
        
        axes[i, j - 1].set_xticks(np.arange(0, 202, 50))
        axes[i, j - 1].set_xticklabels(['-0.5', '-0.25', '0', '0.25', '0.5'])
        
        axes[i, j - 1].set_yticks(np.arange(0, 202, 50))
        axes[i, j - 1].set_yticklabels(['-0.5', '-0.25', '0', '0.25', '0.5'])
    
        IMG = normalized_images[j].values - normalized_images[i].values
    
        pcm = axes[i,j - 1].pcolor(IMG, norm=colors.SymLogNorm(linthresh=0.01, linscale=0.01,
                                                vmin=np.min(IMG),
                                                vmax=np.max(IMG)), cmap=plt.cm.Spectral_r)
    
        divider = make_axes_locatable(axes[i, j-1])
        cax = divider.append_axes("right", size="7%", pad="4%")

        cb = plt.colorbar(pcm, cax=cax)
        tick_locator = ticker.MaxNLocator(nbins=3)
        cb.locator = tick_locator
        cb.update_ticks()

plt.savefig('difference_of_analytic_fields.png', dpi=120)

plt.show()
```

```python
fig, axes = plt.subplots(3, 3, sharex=True, sharey=True, squeeze=False, figsize=(18, 18))


for i in range(3):
    for j in range(3):
        
        axes[i, j].set_title("ELTE field - (%d) analytic field" % (i*3 + j + 1))
        
        axes[i, j].set_visible(True)
        
        axes[i, j].set_xticks(np.arange(0, 202, 50))
        axes[i, j].set_xticklabels(['-0.5', '-0.25', '0', '0.25', '0.5'])
        
        axes[i, j].set_yticks(np.arange(0, 202, 50))
        axes[i, j].set_yticklabels(['-0.5', '-0.25', '0', '0.25', '0.5'])
    
        IMG = img.values - normalized_images[i*3 + j].values
    
        pcm = axes[i,j].pcolor(IMG, norm=colors.SymLogNorm(linthresh=1, linscale=1,
                                                vmin=np.min(IMG),
                                                vmax=np.max(IMG)), cmap=plt.cm.Spectral_r)
    
        divider = make_axes_locatable(axes[i, j])
        cax = divider.append_axes("right", size="7%", pad="4%")

        cb = plt.colorbar(pcm, cax=cax)
        tick_locator = ticker.MaxNLocator(nbins=7)
        cb.locator = tick_locator
        cb.update_ticks()

plt.savefig('difference_of_analytic_field_and_elte_field.png', dpi=120)

plt.show()
```
